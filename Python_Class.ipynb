````python
def add(x, y=20):
    # Function definition
    print('x: ', x)
    print('y: ', y)
    z = x + y
    return z

print('Start...')
a = 10
c = add(a) # Function calling / Function call
print('First: ', c)

a = 20
b = 40
c = add(a, b) # Function calling / Function call
print('Second: ', c)
print('End...')
````

````python
# *args
def add(*args):
    # Function definition
    print('type of: ', type(args))
    print('value of : ', args)
    z = sum(args)
    return z

c = add(10, 20, 30, 40, 50, 60)
print('c: ', c)
````

````python
# *kwargs = keyword arguments
def add(**kwargs):
    # Function definition
    print('type of: ', type(kwargs))
    print('value of : ', kwargs)
    z = kwargs['a'] + kwargs['b'] + kwargs['c'] + kwargs['d'] + kwargs['e'] + kwargs['f']
    return z


c = add(a=10, b=20, c=30, d=40, e=50, f=60)
print('c: ', c)
````

````python
def add(*args, **kwargs):
    # Function definition
    print('type of args: ', type(args))
    print('value of args: ', args)
    print('type of kwargs: ', type(kwargs))
    print('value of kwargs: ', kwargs)

    z = sum(args) + kwargs['c'] + kwargs['d']
    return z


c = add(10, 20, c=30, d=40)
print('c: ', c)
````

````python
s = 'thequickbrownfoxjumpsoverthelazydog'
print('String: ', s)

char_freq = {}
for char in s:
    if char not in char_freq:
        char_freq[char] = 0 # Assigning 0 to new char
    char_freq[char] = char_freq[char] + 1 # Increment by 1 for existing char in the dictionary
char_freq
````

````python
# Unfold / decode the decorator

def uppercase_decorator(main_function): # function as an argument
    def wrapper(): # child / nested function
        result = main_function()
        make_uppercase = result.upper()
        return make_uppercase
    return wrapper  # function returning a function

def say_hello():
    return 'Hello'

alias_wrapper = uppercase_decorator(say_hello)
res = alias_wrapper()
res
````

````python
def uppercase_decorator(function): # function as an argument
    def wrapper(): # child / nested function
        result = function()
        make_uppercase = result.upper()
        return make_uppercase
    return wrapper  # function returning a function


def split_string(function): # function as an argument
    def wrapper(): # child / nested function
        result = function()
        splitted_string = result.split()
        return splitted_string
    return wrapper  # function returning a function


@split_string
@uppercase_decorator
def say_hello():
    return 'Hello World'

say_hello()
````

````python
def args_decorator(function): # function as an argument
    def wrapper(*args, **kwargs): # child / nested function
        print('args: ', args)
        print('kwargs: ', kwargs)
        res = function(args, kwargs)
        res = res - kwargs['y1'] - kwargs['y2']
        return res
    return wrapper  # function returning a function

@args_decorator
def do_add(args, kwargs):
    return sum(args)

do_add(10, 20, 30, 40, y1=50, y2=60)

````

````python
class Parent: # Parent / Base / Super Class
    def __init__(self):
        print('Parent Constructor')

    def parent_method(self):
        print('Parent Method')

# Inheritance - Parent properties ---> Child class.
class Child(Parent): # Child / Derived / Sub Class

    def __init__(self):
        super().__init__() # Parent.__init__(self)
        print('Child Constructor')

    def child_method(self):
        print('Child Method')

# parent = Parent()
# parent.parent_method()

child = Child()
child.child_method()
child.parent_method()
````

````python
class Father: # Parent / Base / Super Class
    def __init__(self):
        print('Father Constructor')

    def father_method(self):
        print('Father Method')

    def common_method(self):
        print('Common(father) Method')

# Inheritance - Parent properties ---> Child class.
class Mother:

    def __init__(self):
        print('Mother Constructor')

    def mother_method(self):
        print('Mother Method')

    def common_method(self):
        print('Common(mother) Method')

# Multiple Inheritance:
class Son(Father, Mother):

    def __init__(self):
        super().__init__()
        print('Son Constructor')

    def son_method(self):
        print('Son Method')

# parent = Parent()
# parent.parent_method()

son = Son()
son.son_method()
son.father_method()
son.mother_method()
son.common_method()
````

````python
def array_equal(a, b):
    # Step 1 = element-wise
    # Step 2 = True / False
    if len(a) != len(b):
        return False

    len_ = len(a)
    bool_list = []
    for index in range(0, len_):
        condition = a[index] == b[index]
        bool_list.append(condition)

    if False in bool_list:
        return False
    return True



a = [1, 2, 3, 4]
b = [2, 2, 1, 4]
array_equal(a, b)
````

````python

# What is the total sales in year 2021 in each region/sector? (product-wise)

# Year, Region, Product, Sales
# 2021, Area1 , A,       10000
# 2021, Area1,  A,       2000
# 2021, Area2,  A,       300
# 2021, Area1,  B,       300
# 2022, Area1,  B,       300
````

````python
df2 = pd.DataFrame({
    'Year': [2021, 2021, 2021, 2022],
    'Region': ['Area1', 'Area1', 'Area2', 'Area1'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [10000, 2000, 300, 3000],
})
df2
````

````python
df2 = df2[df2['Year'] == 2021]
df2
````

````python
df2.pivot_table(index=['Region'],
                columns='Product',
                values='Sales',
                aggfunc='sum',
                fill_value=0)
````

````python
df2.pivot_table(index=['Region'],
                columns='Product',
                values='Sales',
                aggfunc='sum',
                fill_value=0,
                margins=True)
````
